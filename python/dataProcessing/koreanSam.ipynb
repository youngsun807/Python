```python
import re
for i in range(len(df)):
    df.iloc[i, 0] = re.sub('-ㄴ', '', df.iloc[i, 0])
    df.iloc[i, 1] = re.sub("ː", '', df.iloc[i, 1])
    df.iloc[i, 0] = re.sub('-', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('-ㄹ', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('-ㅂ', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('-ㅁ', '', df.iloc[i, 0])
```
