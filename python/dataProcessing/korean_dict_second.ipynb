```python
import pandas as pd
import numpy as np
import re

df = pd.read_excel("C:/0.ITStudy/0.dataset/다.xls", sheet_name='Sheet0')
df.drop(['동형어 번호', '구분', '뜻풀이', '용례'], axis=1, inplace=True)
df.fillna('', inplace=True)
```

```python
import re
for i in range(len(df)):
    df.iloc[i, 0] = re.sub('-ㄴ', '', df.iloc[i, 0])
    df.iloc[i, 2] = re.sub("ː", '', df.iloc[i, 2])
    df.iloc[i, 0] = re.sub('-', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('-ㄹ', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('-ㅂ', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('-ㅁ', '', df.iloc[i, 0])
```

```python
for i in range(len(df)):
    df.iloc[i, 0] = re.sub('ㄹ', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('ㅂ', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('ㅁ', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('-', '', df.iloc[i, 0])
    df.iloc[i, 0] = re.sub('ㄴ', '', df.iloc[i, 0])
```

```python
unique_list = []
no = []
for i in range(len(df)):
    if df.loc[i,'표제어'] in unique_list:
        no.append(i)
        continue
    else:
        unique_list.append(df.loc[i,'표제어'])
print(no)
```

```python
df.drop(no, axis=0, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
for i in range(len(df)):
    if '/' in df.loc[i,'발음']:
        pron = df.loc[i,'발음'].split('/')
        df.loc[i,'발음'] = ''
        for j in pron:
            if df.loc[i,'표제어']==j:
                continue
            else:
                df.loc[i,'발음'] = df.loc[i,'발음']+'/'+j
        continue
    else:
        if df.loc[i,'표제어']==df.loc[i,'발음']:
             df.loc[i,'발음'] = ''
        else:
            continue
```
